```python
import pandas as pd
```

```python
# Specify values for each column.
df = pd.DataFrame(
{"a" : [4 ,5, 6],
"b" : [7, 8, 9],
"c" : [10, 11, 12]},index = [1, 2, 3])
df
```

```python
# Specify values for each row.
df1 = pd.DataFrame(
[[4, 7, 10],
[5, 8, 11],
[6, 9, 12]],
index=[1, 2, 3],
columns=['a', 'b', 'c'])
df1
```

```python
# Create DataFrame with a MultiIndex
df2 = pd.DataFrame(
{"a" : [4 ,5, 6],
"b" : [7, 8, 9],
"c" : [10, 11, 12]
}, 
index = pd.MultiIndex.from_tuples([('d',1),('j',2),('e',3)],names=['n','v']))
df2
```

# Reshaping Data

â€“ Change the layout of a data set

```python
# Gather columns into rows.
print(df2)
pd.melt(df2)
```

```python
# Most pandas methods return a DataFrame so that another pandas method can be applied to the result. 
#This improves readability of code.
df3 = (pd.melt(df2).rename(
    columns={'variable' : 'var','value' : 'val'}).query('val >= 10'))
df3
```

```python
# Spread rows into columns.
df3.pivot(columns='var', values='val')
```

```python
# Append rows of DataFrames
print(df1)
print(df2)
pd.concat([df1,df2])
```

```python
#Append columns of DataFrames
pd.concat([df1,df2],axis=0)
```
