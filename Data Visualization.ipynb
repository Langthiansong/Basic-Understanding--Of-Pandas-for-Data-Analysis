## Adjusting plot

#### 1. Color to represent our figure

```python
#Adjust plot
fig = plt.figure()
plt.plot(x,np.sin(x - 0), color='blue') # color name
plt.plot(x,np.sin(x - 1), color='m') # short color code (rgbcmyk)
plt.plot(x,np.sin(x - 2), color='0.9') # gray scale between 0 and 1
plt.plot(x,np.sin(x - 3), color='#FFDD44') # hex code
plt.plot(x,np.sin(x - 4), color=(1.0,0.5,0.3)) # rgb tupple values 0 to 1
plt.plot(x,np.sin(x - 5), color='chartreuse') # all html color names supported
```

```python
fig.savefig('multicolor_sin.png')
```

```python
from IPython.display import Image
Image('multicolor_sin.png')
```

#### 2. Different types of lines for representation

```python
plt.plot(x, x+0,linestyle='solid')
plt.plot(x, x+1,linestyle='dashed')
plt.plot(x, x+2,linestyle='dashdot')
plt.plot(x, x+3,linestyle='dotted')

# using sybol
plt.plot(x, x+4,linestyle='-')
plt.plot(x, x+5,linestyle='--')
plt.plot(x, x+6,linestyle='-.')
plt.plot(x, x+7,linestyle=':')
```

#### 3. Combing color and type of line for representation

```python
plt.plot(x, x+0,'-g')
plt.plot(x, x+1,'--c')
plt.plot(x, x+2,'-.k')
plt.plot(x, x+3,':r')
```

#### 4. Adding limit to x and y axis

```python
plt.plot(x,np.sin(x))
plt.xlim(-1,11)
```

```python
plt.plot(x,np.sin(x))
plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
```

```python
# Alternate of applying limit using axis()
plt.plot(x,np.sin(x))
plt.axis([-1,11,-1.5,1.5])  # axis() parameter passed then Returns - xmin, xmax, ymin, ymax : float    The axis limits.
```

```python
plt.plot(x,np.sin(x))
plt.axis('tight') # axis() values passed = on, off, equal, scaled, tight, auto, normal, image, square
```

```python
plt.plot(x,np.sin(x))
plt.axis('equal')
```

#### 5. Adding labels

```python
plt.plot(x,np.sin(x))
plt.title('A sin Curve') # Add heading or title to graph
plt.xlabel('X - axis') # Add name to x axis
plt.ylabel('Y - axis') # Add name to y axis
```

#### 6. Combinig color, type, labels and limits of a line putting all together

```python
plt.plot(x,np.sin(x), '-g', label='sin(x)')
plt.plot(x,np.cos(x), ':m', label='sin(x)')
plt.plot(x,np.log(x), '-.r', label='sin(x)')
plt.axis('equal')
plt.legend() # Implement combination of all
```

```python
# Alternate of implementing all together
ax = plt.axes()
ax.plot(x,np.sin(x))
ax.set(xlim=(0,10), ylim=(-2,2),xlabel = 'x',ylabel='sin(x)', title='A Simple Plot')
# set implement all the attributes
```
